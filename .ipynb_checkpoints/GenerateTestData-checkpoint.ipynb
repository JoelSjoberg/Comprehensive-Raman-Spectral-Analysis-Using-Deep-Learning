```python
# Data generator
from Scripts.generator import train_generator
import numpy as np

import matplotlib.pyplot as plt
```

```python
number_of_spectra = 10000
path = "data/Generated_data/"
```

```python
# Choose parameters

spectrum_length = 1024
peak_widths = [10, 300]

X, y = [], []
np.random.seed(0)
for i in train_generator(length = spectrum_length,
                         batch_size = number_of_spectra,
                        min_width = peak_widths[0],
                        max_width = peak_widths[1],
                        increase_noise = True):
    X = i[0]
    y = i[1]
    break

    
ix = 1
plt.figure(figsize = (15, 9))
plt.plot(X[ix])
plt.show()

plt.figure(figsize = (15, 9))
# Baseline
plt.plot(y[0][ix])
# Cosmic rays
plt.plot(y[1][ix])
# Noise
plt.plot(y[2][ix])
# Peaks
plt.plot(y[3][ix])
plt.show()

np.save(path+"X_"+str(spectrum_length) +"_"+ str(peak_widths[0]) + "_" +str(peak_widths[1]),
        X)
np.save(path+"y_baseline_"+str(spectrum_length) +"_"+ str(peak_widths[0]) + "_" + str(peak_widths[1]),
        y[0])
np.save(path+"y_cosmic_rays_"+str(spectrum_length) +"_"+ str(peak_widths[0]) + "_" + str(peak_widths[1]),
        y[1])
np.save(path+"y_noise_"+str(spectrum_length) +"_"+ str(peak_widths[0]) + "_" + str(peak_widths[1]),
        y[2])
np.save(path+"y_peaks_"+str(spectrum_length) +"_"+ str(peak_widths[0]) + "_" + str(peak_widths[1]),
        y[3])
```
