```python
# For Figure 4

inp_color = "Black"
p_color = "Green"
bl_color = "Blue"
cr_color = "Purple"
n_color = "Red"
labels = ["a", "b", "c", "d"]

np.random.seed(0)

lengths = [800, 1000, 2000, 3000]
min_widths = [16, 5, 32, 50]
max_widths = [100, 95, 300, 400]

plt.rcParams.update({'font.size': 18})

#plt.figure(figsize=(5, 5))
cr = 0

for i in range(4):
    bl, cr, n, p = generate_spectrum(lengths[i], min_widths[i], max_widths[i])
    
    spectrum = bl+cr+n+p
    
    maxim = np.max(spectrum)
    spectrum = spectrum / maxim
    bl = bl/maxim
    cr = cr/maxim
    n = n/maxim
    p = p/maxim
    
    fig, axs = plt.subplots(5, sharex=True, sharey=True)
    
    axs[0].plot(spectrum, color = inp_color)
    axs[1].plot(bl , color = bl_color)
    axs[2].plot(cr , color = cr_color)
    axs[3].plot(n , color = n_color)
    axs[4].plot(p, color = p_color)
    
    #plt.gca().set_yticklabels([-1, 1])
    #plt.tick_params(left = False)
    plt.savefig("Figures/Examples/generated_spectrum_1024_+"+str(i)+"_+.png",
                bbox_inches='tight',
                pad_inches=0.5)
    plt.show()
```

```python
# Figure 3, show the order in which each component is extracted and removed

stan_model = make_standard()
stan_model.load_weights(filepath = "Models/standardModel.h5")

np.random.seed(3)
inp_color = "Black"
p_color = "Green"
bl_color = "Blue"
cr_color = "Purple"
n_color = "Red"
labels = ["a", "b", "c", "d"]

lengths = [800, 1000, 2000, 3500]
min_widths = [16, 5, 32, 50]
max_widths = [100, 95, 400, 300]

plt.rcParams.update({'font.size': 23})

for i in range(1):
    bl, cr, n, p = generate_spectrum(lengths[i], min_widths[i], max_widths[i])
    spectrum = bl+cr+n+p
    
    maxim = np.max(spectrum)
    spectrum = spectrum / maxim
    bl = bl/maxim
    cr = cr/maxim
    n = n/maxim
    p = p/maxim
    
    preds = stan_model.predict(spectrum.reshape(1, -1, 1))
    
    p_bl = preds[0][0]
    p_cr = preds[1][0]
    p_n = preds[2][0]
    p_p = preds[3][0]
    
    
    fig = plt.figure(figsize=(2, 5))
    
    plt.plot(spectrum, np.arange(len(spectrum)), color = inp_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/standard_inp_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_bl, np.arange(len(spectrum)), color = bl_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/standard_bl_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_cr, np.arange(len(spectrum)), color = cr_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/standard_cr_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_n, np.arange(len(spectrum)), color = n_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/standard_n_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_p, np.arange(len(spectrum)), color = p_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/standard_p_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
```

```python
# Figure 3, show the order in which each component is extracted and removed

ensemble_model = make_ensemble()
ensemble_model.load_weights(filepath = "Models/ensembleModelRes.h5")

np.random.seed(4)
inp_color = "Black"
p_color = "Green"
bl_color = "Blue"
cr_color = "Purple"
n_color = "Red"
labels = ["a", "b", "c", "d"]

lengths = [800, 1000, 2000, 3500]
min_widths = [16, 5, 32, 50]
max_widths = [100, 95, 400, 300]

plt.rcParams.update({'font.size': 23})

for i in range(1):
    bl, cr, n, p = generate_spectrum(lengths[i], min_widths[i], max_widths[i])
    spectrum = bl+cr+n+p
    
    maxim = np.max(spectrum)
    spectrum = spectrum / maxim
    bl = bl/maxim
    cr = cr/maxim
    n = n/maxim
    p = p/maxim
    
    preds = ensemble_model.predict(spectrum.reshape(1, -1, 1))
    
    p_bl = preds[0][0]
    p_cr = preds[1][0]
    p_n = preds[2][0]
    p_p = preds[3][0]
    
    
    fig = plt.figure(figsize=(2, 5))
    
    plt.plot(spectrum, np.arange(len(spectrum)), color = inp_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/hierarchical_inp_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_bl, np.arange(len(spectrum)), color = bl_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/hierarchical_bl_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_cr, np.arange(len(spectrum)), color = cr_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/hierarchical_cr_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(spectrum - p_bl - p_cr, np.arange(len(spectrum)), color = inp_color, alpha = 0.35)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/hierarchical_inp_minus_bl_cr_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_n, np.arange(len(spectrum)), color = n_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/hierarchical_n_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
    
    fig = plt.figure(figsize=(2, 5))
    plt.plot(p_p, np.arange(len(spectrum)), color = p_color, alpha = 1)
    plt.xlim([-0.2, 1.2])
    plt.axis("off")
    plt.savefig("Figures/Examples/hierarchical_p_"+str(i)+".png",
                bbox_inches='tight',
                dpi = 300,
                transparent = True,
                pad_inches=0.5)
    plt.show()
```
